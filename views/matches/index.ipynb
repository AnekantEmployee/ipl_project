```python
won_by_team1 = len(df[df['winner'] == 'Royal Challengers Bangalore'])
won_by_team2 = len(df[df['winner'] == 'Kolkata Knight Riders'])
drawn = len(df[df['winner'].isnull()])

print(won_by_team1)
print(won_by_team2)
print(drawn)
```

```python
home_won_by_team1 = len(df[(df['winner'] == 'Royal Challengers Bangalore') & (df['city'] == 'Bangalore')])
home_loss_by_team1 = len(df[(df['winner'] == 'Kolkata Knight Riders') & (df['city'] == 'Bangalore')])
away_won_by_team1 = len(df[(df['winner'] == 'Royal Challengers Bangalore') & (df['city'] != 'Bangalore')])
away_loss_by_team1 = len(df[(df['winner'] == 'Kolkata Knight Riders') & (df['city'] != 'Bangalore')])

print(home_won_by_team1)
print(home_loss_by_team1)
print(away_won_by_team1)
print(away_loss_by_team1)
```
